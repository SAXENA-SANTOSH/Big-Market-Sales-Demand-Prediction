```python
item_outlet_identifier = data[["Item_Identifier","Outlet_Identifier"]]
```

```python
X = data.drop(["Item_Identifier","Outlet_Identifier","Item_Outlet_Sales"],axis=1)
```

```python
y = data["Item_Outlet_Sales"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
pred = reg.predict(X_test)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rad_reg = RandomForestRegressor()
```

```python
rad_reg.fit(X_train,y_train)
```

```python
rad_reg.score(X_train,y_train)
```

```python
rad_reg.score(X_test,y_test)
```

```python
coef6 = pd.Series(rad_reg.feature_importances_, X.columns).sort_values(ascending=False)
coef6.plot(kind='bar', title='Feature Importances')
```
